```python
google  = Company("GOOGL")
filings_google = google.get_filings(form="10-K")
tenk_google = [x.obj() for x in filings_google.latest(29)]
```

```python
str_google = ""
```

```python
str_MSFT =""
```

```python
str_nvidia = ""
```

```python
str_amazon = ""
```

```python
for i in range (len(tenk_google)):
    for tenk in tenk_google[0].items:
        if(tenk_google[i][tenk] is not None):

         str_google+= tenk_google[i][tenk]
```

```python
# Initialize the string variable and concatenete the strings
for i in range (len(tenk_all)):
    for tenk in tenk_all[0].items:
        if(tenk_all[i][tenk] is not None):

         str_MSFT+= tenk_all[i][tenk]



```

```python
for i in range (len(tenk_nvidia)):
    for tenk in tenk_nvidia[0].items:
        if(tenk_nvidia[i][tenk] is not None):

         str_nvidia+= tenk_nvidia[i][tenk]
```

```python
for i in range (len(tenk_amazon)):
    for tenk in tenk_amazon[0].items:
        if(tenk_amazon[i][tenk] is not None):

         str_amazon+= tenk_amazon[i][tenk]
```

```python
with open("google.txt", "w", encoding="utf-8") as file:
    file.write(str_google)

print("File 'google.txt' has been created successfully.")
```

```python
# Write str_MSFT to a text file with UTF-8 encoding
with open("MSFT.txt", "w", encoding="utf-8") as file:
    file.write(str_MSFT)

print("File 'output.txt' has been created successfully.")
```
